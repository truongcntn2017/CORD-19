## 3.1 Linguistic Features 

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(df['title'][0])
```

```python
df['title'][0]
```

```python
for token in doc:
    if token.pos_ == 'ADJ' or token.pos_ == 'NOUN':
        print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop)
```
